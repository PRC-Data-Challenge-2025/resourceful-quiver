```python
import pandas as pd
from resqu import data

from tqdm.contrib.concurrent import process_map

import functools
import matplotlib.pyplot as plt
```

```python
def get_acars_stat(fid):
    df = data.get_raw_df(fid)
    acars_df = df[df.source=="acars"]
    return acars_df

result = process_map(get_acars_stat, data.flight_list_rank.flight_id, chunksize=8)
acars_df = pd.concat(result)
```
