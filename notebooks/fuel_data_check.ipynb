```python
from resqu import data, plotter
import pandas as pd
import matplotlib.pyplot as plt

import ipywidgets as widgets
```

```python
# fuel_dup_mask = data.fuel_train_data.duplicated(subset=["start", "end"], keep=False)
# fids = data.fuel_train_data.flight_id[fuel_dup_mask].unique().tolist()

# fuel_dup_mask = data.fuel_rank_data.duplicated(subset=["start", "end"], keep=False)
# fids = data.fuel_rank_data.flight_id[fuel_dup_mask].unique().tolist()

fuel_dup_mask = data.fuel_final_data.duplicated(subset=["start", "end"], keep=False)
fids = data.fuel_final_data.flight_id[fuel_dup_mask].unique().tolist()
```

```python
def plot_and_print(i):
    fid = fids[i]
    plotter.vis_flight_widget(fid)
    print(data.get_fuel_id(fid))
```

```python
widgets.interact(plot_and_print, i=widgets.IntSlider(min=0, max=len(fids)-1))
```
